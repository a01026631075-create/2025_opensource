```python
import json
import os

DB_PATH = "todo_data.json"

def load_todos():
    if os.path.exists(DB_PATH):
        with open(DB_PATH, "r", encoding="utf-8") as f:
            return json.load(f)
    return []

def save_todos(todos):
    with open(DB_PATH, "w", encoding="utf-8") as f:
        json.dump(todos, f, ensure_ascii=False, indent=2)

def list_todos(todos):
    if not todos:
        print("할 일이 없습니다.")
        return
    for i, item in enumerate(todos, 1):
        status = "[x]" if item.get("done") else "[ ]"
        print(f"{i}. {status} {item['task']}")

def main():
    todos = load_todos()
    print("간단 Todo 리스트 (명령어: add, remove, toggle, list, save, exit)")
    while True:
        cmd = input("> ").strip().split(maxsplit=1)
        if not cmd:
            continue
        action = cmd[0].lower()
        arg = cmd[1] if len(cmd) > 1 else ""
        if action == "add":
            if not arg:
                arg = input("할 일 입력: ").strip()
            if arg:
                todos.append({"task": arg, "done": False})
                print("추가되었습니다.")
        elif action == "remove":
            try:
                idx = int(arg) - 1
                removed = todos.pop(idx)
                print(f"삭제: {removed['task']}")
            except Exception:
                print("유효한 번호를 입력하세요. 예: remove 2")
        elif action == "toggle":
            try:
                idx = int(arg) - 1
                todos[idx]["done"] = not todos[idx].get("done", False)
                print("상태 변경됨.")
            except Exception:
                print("유효한 번호를 입력하세요. 예: toggle 1")
        elif action == "list":
            list_todos(todos)
        elif action == "save":
            save_todos(todos)
            print("저장 완료.")
        elif action == "exit":
            save_todos(todos)
            print("저장 후 종료합니다.")
            break
        else:
            print("알 수 없는 명령어입니다. (add/remove/toggle/list/save/exit)")

if __name__ == "__main__":
    main()
```
