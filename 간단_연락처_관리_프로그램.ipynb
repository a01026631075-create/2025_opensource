```python
import json
import os

FILE = "contacts.json"

def load_contacts():
    if os.path.exists(FILE):
        with open(FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def save_contacts(contacts):
    with open(FILE, "w", encoding="utf-8") as f:
        json.dump(contacts, f, ensure_ascii=False, indent=2)

def main():
    contacts = load_contacts()
    print("연락처 관리기 (add/find/remove/list/save/exit)")
    while True:
        cmd_line = input("> ").strip()
        if not cmd_line:
            continue
        parts = cmd_line.split(maxsplit=2)
        cmd = parts[0].lower()
        if cmd == "add":
            if len(parts) >= 3:
                name, number = parts[1], parts[2]
            else:
                name = input("이름: ").strip()
                number = input("번호: ").strip()
            contacts[name] = number
            print(f"{name} 추가/업데이트됨.")
        elif cmd == "find":
            if len(parts) >= 2:
                name = parts[1]
            else:
                name = input("검색할 이름: ").strip()
            num = contacts.get(name)
            if num:
                print(f"{name}: {num}")
            else:
                print("찾을 수 없음.")
        elif cmd == "remove":
            if len(parts) >= 2:
                name = parts[1]
            else:
                name = input("삭제할 이름: ").strip()
            if name in contacts:
                del contacts[name]
                print(f"{name} 삭제됨.")
            else:
                print("없는 이름입니다.")
        elif cmd == "list":
            if not contacts:
                print("연락처가 비어있음.")
            else:
                for name, num in contacts.items():
                    print(f"{name}: {num}")
        elif cmd == "save":
            save_contacts(contacts)
            print("저장 완료.")
        elif cmd == "exit":
            save_contacts(contacts)
            print("저장 후 종료합니다.")
            break
        else:
            print("명령어: add/find/remove/list/save/exit")

if __name__ == "__main__":
    main()
```
