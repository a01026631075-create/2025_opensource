```python
import random

def generate_lotto():
    return sorted(random.sample(range(1, 46), 6))

def main():
    print("로또 번호 생성기")
    try:
        n = int(input("몇 세트 생성할까요? (기본 1): ") or "1")
    except ValueError:
        n = 1
    for i in range(1, n+1):
        print(f"세트 {i}: {generate_lotto()}")

if __name__ == "__main__":
    main()
```
