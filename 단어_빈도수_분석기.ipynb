```python
import re
from collections import Counter

def normalize_text(text):
    text = text.lower()
    return re.sub(r'[^0-9a-zA-ㄱ-힣\s]', ' ', text)

def read_input():
    print("문장을 입력하세요. (빈 줄 입력 시 종료)")
    lines = []
    while True:
        line = input()
        if line.strip() == '':
            break
        lines.append(line)
    return "\n".join(lines)

def word_freq(text):
    norm = normalize_text(text)
    words = [w for w in norm.split() if w]
    return Counter(words)

def main():
    text = read_input()
    if not text.strip():
        print("입력 없음. 종료.")
        return
    freq = word_freq(text)
    print("\n단어 빈도수 (내림차순):")
    for word, count in freq.most_common():
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```
